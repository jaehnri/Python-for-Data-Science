```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

_____________

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

_____________

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_ylabel('y')
ax.set_xlabel('x')
ax.set_title('title')
```

_____________

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])

ax1.plot(x,y)
ax2.plot(x,y)

fig.tight_layout()
```

_____________

```python
fig = plt.figure()

eixo1 = fig.add_axes([0,0,1,1])
eixo2 = fig.add_axes([0.2,0.5,.4,.4])

eixo1.plot(x,z)
eixo1.set_xlim(0,100)
eixo1.set_ylim(0,10000)

eixo2.plot(x,y)
eixo2.set_xlim([20,22])
eixo2.set_ylim([30,50])
eixo2.set_title('Zoom')
```

_____________

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y, 'b--', lw=3)

axes[1].plot(x,z,'r-', lw=3)
```

_____________

```python
fig,axes = plt.subplots(nrows=1, ncols=2,figsize=(8,2))

axes[0].plot(x,y, 'b--', lw=3)
axes[1].plot(x,z,'r-', lw=3)

plt.tight_layout()
```
