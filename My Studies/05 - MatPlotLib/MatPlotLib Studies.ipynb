```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
x
```

```python
y = x**2
```

```python
y
```

```python
# FUNCTIONAL
```

```python
plt.plot(x, y, 'r-')
plt.xlabel('t(s)')
plt.ylabel('s(m)')
plt.title('My First MatPlotLib Graph')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

# OO Method
____________________________________________________________

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X LABEL')
axes.set_ylabel('Y LABEL')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes1.spines['bottom'].set_color('red')
axes1.spines['top'].set_color('red') 
axes1.spines['right'].set_color('red')
axes1.spines['left'].set_color('red')

axes2 = fig.add_axes([0.5, 0, 0.5, 0.5])    #(left, bottom, width and height)
axes2.spines['bottom'].set_color('blue')
axes2.spines['top'].set_color('blue') 
axes2.spines['right'].set_color('blue')
axes2.spines['left'].set_color('blue')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,0.9,0.9])
axes1.spines['bottom'].set_color('red')
axes1.spines['top'].set_color('red') 
axes1.spines['right'].set_color('red')
axes1.spines['left'].set_color('red')
axes1.set_xlabel('X LABEL')
axes1.set_ylabel('Y LABEL')
axes1.set_title('X/Y PLOT', fontsize= 30)

axes2 = fig.add_axes([1, 0, 0.9, 0.9])    #(left, bottom, width and height)
axes2.spines['bottom'].set_color('blue')
axes2.spines['top'].set_color('blue') 
axes2.spines['right'].set_color('blue')
axes2.spines['left'].set_color('blue')
axes2.set_xlabel('Y LABEL')
axes2.set_ylabel('X LABEL')
axes2.set_title('Y/X PLOT', fontsize= 30)

axes1.plot(x,y, 'r--')
axes2.plot(y,x, 'b*-')
```

# Second Lecture
_____________

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout()
```

# Figure Size and DPI

```python
fig,axes = plt.subplots(nrows = 2, ncols = 1, figsize=(8,4))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
figura = plt.figure()                                                #instancia uma nova figura

eixo = figura.add_axes([0,0,1,1])                                    #instancia novos eixos
eixo.plot(x,x**2, label='X SQUARED')                                 #implanta uma funcao
eixo.plot(x,x**3, label='X CUBED')                                   #implanta outra funcao utilizando os mesmos eixos

eixo.legend(loc=4)
```

# Plot Range
_____________

```python
fig = plt.figure()

eixo = fig.add_axes([0,0,1,1])
eixo.plot(x,y, color='#BB15CF', linewidth=0.8, alpha=1, linestyle='-', marker='+', markersize=9,
          markeredgecolor='black')   

eixo.set_xlim([0,1])
eixo.set_ylim([0,2])
```
